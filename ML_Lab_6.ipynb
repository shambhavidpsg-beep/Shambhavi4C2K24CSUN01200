collections.Counter is a high-performance Python tool for counting the frequency of elements in an iterable (like lists, strings, or tuples). It returns a dictionary-like object where items are keys and their counts are values. It's a subclass of dict that provides a fast, concise way to tally data without manual loops

```python
import numpy as np
from collections import Counter
```

Thе KNN algorithm operates on the principle of similarity where it predicts the label or value of a new data point by considering the labels or values of its K nearest neighbors in the training dataset.

```python
def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((np.array(point1) - np.array(point2))**2))
```

```python
def knn_predict(training_data, training_labels, test_point, k):
    distances = []
    for i in range(len(training_data)):
        dist = euclidean_distance(test_point, training_data[i])
        distances.append((dist, training_labels[i]))
    distances.sort(key=lambda x: x[0])
    k_nearest_labels = [label for _, label in distances[:k]]
    return Counter(k_nearest_labels).most_common(1)[0][0]
```

```python
training_data = [[1, 2], [2, 3], [3, 4], [6, 7], [7, 8], [8,9], [9,10], [10,11], [12,13], [14,15], [15,16]]
training_labels = ['A', 'A', 'A', 'B', 'B', 'A', 'A','B', 'B', 'B', 'B', 'B']
test_point = [7, 8]
k = 5
```

```python
prediction = knn_predict(training_data, training_labels, test_point, k)
print(prediction)
```
