```python
#Data Normalizaton
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame([
    [180000, 110, 18.9, 1400],
    [360000, 905, 23.4, 1800],
    [230000, 230, 14.0, 1300],
    [60000, 450, 13.5, 1500]
], columns=['Col A', 'Col B', 'Col C', 'Col D'])

print(df)

df.plot(kind='bar')
plt.show()
```

```python
#Maximum Absolute Scaling: It will convert the data in rang of -1 to 1
```

```python
max_scaled = df.copy()

for column in max_scaled.columns:
    max_scaled[column] = max_scaled[column] / max_scaled[column].abs().max()

print(max_scaled)

max_scaled.plot(kind='bar')
plt.show()
```

```python
max_scaled = df.copy()

for column in max_scaled.columns:
    max_scaled[column] = max_scaled[column] / max_scaled[column].abs().max()

print(max_scaled)

max_scaled.plot(kind='line')
plt.show()
```

```python
#Min Max Feature Scaling- it rescales the feature to a hard and fast range of [0,1] by subtracting the minimum value of the feature then dividing by the range.
```

```python
scaled = df.copy()

for column in scaled.columns:
    scaled[column] = (scaled[column] - scaled[column].min()) / (scaled[column].max() - scaled[column].min())

print(scaled)
scaled.plot(kind='bar')
plt.show()
```

```python
#Z Score method- The z-score method often called standardization changes the values in each column so that they have a mean of 0 and a standard deviation of 1.
```

```python
z_scaled = df.copy()

for column in z_scaled.columns:
    z_scaled[column] = (z_scaled[column] - z_scaled[column].mean()) / z_scaled[column].std()

print(z_scaled)

z_scaled.plot(kind='bar')
plt.show()
```
